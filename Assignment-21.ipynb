### 1. What is a probability distribution, exactly ? If the values are meant to be random, how can you predict them at all ?

#### Solution :- A probability distribution is a mathematical function that describes the likelihood of observing different outcomes or events in a random process. In other words, it gives the probabilities of all possible outcomes of a random variable. For example, if you toss a fair coin, the probability distribution of the outcomes would be 0.5 for heads and 0.5 for tails. In this case, the probability distribution is a discrete distribution since there are only two possible outcomes.

#### On the other hand, if you measure the height of a large number of people, the probability distribution of the heights would be a continuous distribution. It would describe the probabilities of observing different heights within a certain range. When we say that the values of a probability distribution are "random," we mean that the outcome of the random process is uncertain or unknown. However, a probability distribution allows us to predict the probabilities of different outcomes, based on our understanding of the underlying process and any available data. For example, we can use the probability distribution of a fair coin to predict that we are likely to observe roughly an equal number of heads and tails over a large number of tosses.

#### So while we cannot predict the exact outcome of a random process, we can use probability distributions to make probabilistic predictions about the likelihood of different outcomes.

### 2. Is there a distinction between true random numbers and pseudo-random numbers, if there is one ? Why are the latter considered “good enough” ?

#### Solution :- There is a distinction between true random numbers and pseudo-random numbers. True random numbers are generated from a physical process that is inherently unpredictable, such as atmospheric noise or radioactive decay. They are truly random in the sense that their values cannot be predicted or reproduced with certainty.

#### In contrast, pseudo-random numbers are generated by an algorithm that uses a deterministic process to generate a sequence of numbers that appear to be random. While the numbers are not truly random, they can be useful in many applications where a source of true randomness is not available or practical.

#### Pseudo-random numbers are considered "good enough" for many applications because they exhibit statistical properties that are similar to true random numbers. In particular, they should have a uniform distribution and be independent of each other. If these properties are met, then the sequence of numbers should be sufficiently unpredictable for most practical purposes.

#### There are, however, some applications where true randomness is essential. For example, in cryptography, true random numbers are used to generate encryption keys, and any predictability in the key generation process can compromise the security of the encryption. In such cases, specialized hardware or external sources of randomness may be used to generate true random numbers.

### 3. What are the two main factors that influence the behaviour of a "normal" probability distribution ?

#### Solution :- The behavior of a normal probability distribution is mainly influenced by two factors: the mean and the standard deviation. The mean (also known as the expected value) of a normal distribution determines its central tendency. It represents the average value of the distribution, and the highest point of the curve is located at the mean. A change in the mean shifts the entire distribution left or right, without changing its shape.

#### The standard deviation of a normal distribution determines its spread or variability. It represents how much the values in the distribution deviate from the mean. The larger the standard deviation, the more spread out the distribution is, and the flatter the curve becomes. A smaller standard deviation results in a narrower, more peaked distribution.

#### Together, the mean and standard deviation fully characterize the shape and location of a normal distribution. In fact, any normal distribution can be uniquely described by its mean and standard deviation, and this makes the normal distribution an important and widely used tool in statistics and data analysis.

#### The standard deviation of the height distribution is about 3 inches (7.6 centimeters) for both males and females. This means that most people fall within three inches of the mean height, and the distribution tapers off gradually as you move further away from the mean. The normal distribution of heights has important practical applications, such as in designing clothes, furniture, and transportation systems that fit a wide range of people. Additionally, the normal distribution is used in statistical inference to make predictions about population characteristics based on a sample of data.

### 5. In the short term, how can you expect a probability distribution to behave ? What do you think will happen as the number of trials grows ?

#### Solution :- In the short term, the behavior of a probability distribution can be somewhat unpredictable and subject to random variation. However, as the number of trials grows, the behavior of the distribution becomes more consistent and predictable.

#### In the short term, the results of individual trials may deviate significantly from the expected values based on the probability distribution. For example, if you flip a fair coin ten times, you might get 7 heads and 3 tails, which is not exactly what you would expect based on the 50/50 probability of heads and tails. However, as you continue to flip the coin more and more times, the actual results will tend to converge toward the expected probability distribution.

#### As the number of trials grows, the distribution of outcomes becomes more tightly clustered around the expected values, and the random variation becomes less significant. This is known as the law of large numbers, which states that the average of a large number of independent and identically distributed random variables approaches the expected value as the sample size increases.

#### In practical terms, this means that if you repeat a random process enough times, the probability distribution of the outcomes will tend to stabilize and become more predictable, allowing you to make more accurate predictions and estimates based on the expected values.
